```python
import pandas as pd
import matplotlib.pyplot as plt

# Agrupar por categoria e somar o total de vendas
ranking_categorias = loja.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)

# Criar o gráfico de barras
plt.figure(figsize=(9, 5))
barras = plt.bar(ranking_categorias.index, ranking_categorias.values, color='#4B9CD3', edgecolor='black')

# Título e rótulos
plt.title('🏆 Faturamento por Categoria de Produto', fontsize=14, fontweight='bold')
plt.xlabel('Categoria', fontsize=11)
plt.ylabel('Faturamento (R$)', fontsize=11)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adicionar os valores em reais acima das barras
for barra in barras:
    valor = barra.get_height()
    texto = f"R$ {valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
    plt.text(barra.get_x() + barra.get_width() / 2, valor + (valor * 0.01), texto,
             ha='center', va='bottom', fontsize=9, fontweight='bold', color='#333333')

# Ajustar o layout
plt.tight_layout()
plt.show()
from google.colab import files
plt.savefig('🏆 Faturamento por Categoria de Produto.png', dpi=300, bbox_inches='tight')

# Exibir o gráfico no Colab
plt.show()
files.download('🏆 Faturamento por Categoria de Produto.png')
plt.savefig('🏆 Faturamento por Categoria de Produto', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Agrupar por produto e somar vendas
ranking_produtos = loja.groupby('Produto')['Preço'].sum().sort_values(ascending=False)

# Número de produtos principais que você quer destacar
top_n = 10

# Separar top N produtos e agrupar o restante em 'Outros'
top_produtos = ranking_produtos[:top_n]
outros_valor = ranking_produtos[top_n:].sum()

# Criar nova série incluindo 'Outros'
valores_pizza = pd.concat([top_produtos, pd.Series({'Outros': outros_valor})])


# Cores
cores = plt.cm.tab20.colors

# Explodir a fatia do produto que mais faturou
explode = [0.1] + [0]*(len(valores_pizza)-1)

# Criar gráfico de pizza
plt.figure(figsize=(9,9))
plt.pie(
    valores_pizza,
    labels=valores_pizza.index,
    autopct=lambda p: f"R$ {p * valores_pizza.sum() / 100:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."),
    colors=cores,
    startangle=90,
    wedgeprops={'edgecolor':'black'},
    explode=explode,
    shadow=True
)

plt.title(f'🏆 Top {top_n} Produtos Mais Vendidos + Outros (R$)', fontsize=14, fontweight='bold')
plt.axis('equal')  # Mantém a pizza redonda

# Salvar o gráfico
nome_arquivo = 'pizza_produtos_outros.png'
plt.savefig(nome_arquivo, dpi=300, bbox_inches='tight')

# Exibir gráfico
plt.show()

# Baixar gráfico
files.download(nome_arquivo)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Estatísticas básicas ---
media_avaliacao = loja['Avaliação da compra'].mean()
mediana_avaliacao = loja['Avaliação da compra'].median()
moda_avaliacao = loja['Avaliação da compra'].mode()[0]

print(f"Média das avaliações: {media_avaliacao:.2f}")
print(f"Mediana das avaliações: {mediana_avaliacao}")
print(f"Moda das avaliações: {moda_avaliacao}")

# --- Contagem de cada nota ---
contagem_notas = loja['Avaliação da compra'].value_counts().sort_index()

# --- Exibir gráfico de barras ---
plt.figure(figsize=(8,5))
plt.bar(contagem_notas.index, contagem_notas.values, color='royalblue', edgecolor='black')
plt.title('Distribuição das Avaliações dos Clientes', fontsize=14, fontweight='bold')
plt.xlabel('Nota de Avaliação')
plt.ylabel('Quantidade de Clientes')
plt.xticks([1,2,3,4,5])
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

# --- Percentual de avaliações positivas (>=4) ---
positivas = (loja['Avaliação da compra'] >= 4).sum()
total = loja['Avaliação da compra'].count()
percentual_positivas = (positivas / total) * 100

print(f"\n📈 Avaliações positivas (4 e 5): {percentual_positivas:.2f}% do total")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files  # Para baixar o arquivo no Colab

# --- Filtrar avaliações positivas (nota >= 4) ---
avaliacoes_positivas = loja[loja['Avaliação da compra'] >= 4]

# --- Agrupar por produto: média e quantidade de avaliações ---
resumo_produtos = avaliacoes_positivas.groupby('Produto')['Avaliação da compra'].agg(['mean','count'])

# --- Selecionar TOP 5 produtos por média de avaliação ---
top5_produtos = resumo_produtos.sort_values('mean', ascending=False).head(5)

# --- Gráfico de bolhas ---
plt.figure(figsize=(10,6))
plt.scatter(
    x=top5_produtos.index,             # nomes dos produtos
    y=top5_produtos['mean'],           # média das avaliações
    s=top5_produtos['mean']*300,       # tamanho das bolhas proporcional à média
    color='mediumseagreen',
    alpha=0.7,
    edgecolors='black'
)

# Adicionar valores das médias acima das bolhas
for i, valor in enumerate(top5_produtos['mean']):
    plt.text(i, valor + 0.05, f"{valor:.2f}", ha='center', fontsize=10, fontweight='bold')

plt.title('🌟 TOP 5 Produtos Mais Bem Avaliados (nota ≥4)', fontsize=14, fontweight='bold')
plt.xlabel('Produto', fontsize=12)
plt.ylabel('Média das Avaliações', fontsize=12)
plt.ylim(4,5.2)  # Limite do eixo y para destacar notas altas
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

# --- Salvar gráfico em .png ---
nome_arquivo = 'top5_produtos_bem_avaliados.png'
plt.savefig(nome_arquivo, dpi=300, bbox_inches='tight')

# --- Exibir gráfico ---
plt.show()

# --- Baixar gráfico automaticamente ---
files.download(nome_arquivo)
```
